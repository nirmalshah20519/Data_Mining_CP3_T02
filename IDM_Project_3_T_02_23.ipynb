```python
matches_df = matches_df.drop(columns=column)
```

```python
rankings = {
    "India" : 1,
    "Australia" : 2,
    "South Africa" : 3,
    "Pakistan" : 4,
    "New Zealand" : 5,
    "England" : 6,
    "Sri Lanka" : 7,
    "Bangladesh" : 8,
    "Afghanistan" : 9,
    "Netherlands" : 14,
}
```

```python
matches_df['team1_rank'] = matches_df['team1'].map(rankings)
matches_df['team2_rank'] = matches_df['team2'].map(rankings)

matches_df
```

```python
matches_df.columns
```

```python
new_order = ['team1','team1_rank', 'team2','team2_rank', 'venue', 'city', 'winner',
       ]
matches_df = matches_df[new_order]
matches_df
```

```python
cleaned = matches_df
cleaned = cleaned.reset_index(drop=True)
cleaned.loc[cleaned['winner'] == cleaned['team2'], 'winning_team']=2
cleaned.loc[cleaned['winner'] == cleaned['team1'], 'winning_team']=1
cleaned = cleaned.drop(['winner'], axis=1)
cleaned['winning_team']=cleaned['winning_team'].astype(int)
cleaned
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

X = cleaned.drop(columns = ['winning_team'])
y = cleaned['winning_team']
# One-Hot Encoding

dfc = pd.get_dummies(X, columns=['team1', 'team2', 'venue', 'city',], prefix='encoded',drop_first=True)

# Print the final DataFrame with label-encoded and one-hot encoded columns
dfc
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
numerical_features = ['team1_rank', 'team2_rank', ]
scaler = StandardScaler()
dfc[numerical_features] = scaler.fit_transform(dfc[numerical_features])
dfc
```

```python
y
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score

# Assuming you have a DataFrame named 'df' with features and labels
# Adjust this based on your actual DataFrame structure


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(dfc, y, test_size=0.2, )



# Define the model
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
# model.add(Dropout(0.5))  # Optional: Dropout for regularization
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
# model.add(Dropout(0.5))  # Optional: Dropout for regularization
model.add(Dense(1, activation='sigmoid'))  # Sigmoid activation for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=8, validation_split=0.2)

# Evaluate the model on the testing set
y_pred = model.predict(X_test)
y_pred_binary = (y_pred > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred_binary)
print("Accuracy:", accuracy)
```
